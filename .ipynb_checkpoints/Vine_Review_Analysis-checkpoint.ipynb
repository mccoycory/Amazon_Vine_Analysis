```python
import pandas as pd
```

```python
df = pd.read_csv(r'C:\Users\corym\OneDrive\Documents\Github\Amazon_Vine_Analysis\vine_table.csv')
```

```python
df_20 = df.loc[df["total_votes"]>20]
```

```python
df_20["% Helpful"] = (df_20["helpful_votes"]/df_20["total_votes"])*100
```

```python
df_helpful = df_20.loc[df_20["% Helpful"]> 50]
```

```python
vine_Y_df = df_helpful.loc[df_helpful["vine"]== "Y"]
vine_Y_df
```

```python
vine_N_df = df_helpful.loc[df_helpful["vine"]== "N"]
vine_N_df
```

```python
star_rating_N_df = vine_N_df.groupby("star_rating").count().reset_index()
```

```python
star_rating_N_df["Percentage of Votes"] = (star_rating_N_df["total_votes"]/star_rating_N_df['total_votes'].sum())*100
```

```python
star_rating_Y_df = vine_Y_df.groupby("star_rating").count().reset_index()
```

```python
star_rating_Y_df["Percentage of Votes"] = (star_rating_Y_df["total_votes"]/star_rating_Y_df['total_votes'].sum())*100
```

```python
star_rating_Y_df.loc['Total']= star_rating_Y_df.sum()
star_rating_Y_df
```

```python
star_rating_N_df.loc['Total']= star_rating_N_df.sum()
star_rating_N_df
```
